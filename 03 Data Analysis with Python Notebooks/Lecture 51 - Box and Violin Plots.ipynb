```python
import numpy as np
from numpy.random import randn
import pandas as pd

from scipy import stats

import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
data1 = randn(100)
data2 = randn(100)
```

```python
sns.boxplot([data1])
```

```python
sns.boxplot([data2])
```

```python
sns.boxplot([data1,data2])
```

```python
#INCluding the outliers within the range of the whiskers
```

```python
sns.boxplot([data1,data2],whis=np.inf)
```

```python
sns.boxplot(data1,orient='h')
```

```python
sns.boxplot(data1,orient='v')
```

```python
#EXAMPLE WHERE THE BOXPLOT WONT GIVE THE FULL PICTURE
```

```python
data1 = stats.norm(0,5).rvs(100)
```

```python
#TWO GAMA DISTRIBUTIONS CONCATINATED
```

```python
data2 = np.concatenate([stats.gamma(5).rvs(50)-1,
                       -1*stats.gamma(5).rvs(50)])


#box plot both data1 and data2


sns.boxplot([data1,data2],whis=np.inf)
```

```python
sns.violinplot([data1,data2])
```

```python
sns.violinplot([data1,data2],bw=0.01)
```
